```python
filename_in = input('CSV file name: ')
```

```python
df_et_a = pd.read_csv(filename_in, header=None)
df_et_a
```

```python
month = input('input month: ')
```

```python
# HeatMap
fig, ax = plt.subplots(figsize=(14,7))         # Sample figsize in inches
heat_map = sns.heatmap(df_et_a, xticklabels=False, yticklabels=False, vmin=-1.5e-5, vmax=2.5e-5, 
                       cbar_kws={'label': '$kg\;m^{-2}\;s^{-1}$'}) #cmap="PiYG"

# Add Title
plt.title('Time Averaged Map of Anomaly of Total evapotranspiration monthly 0.1 deg. \n FLDAS_NOAH01_CP_GL_M \n over 2020-'+month+' \n')

# Display
plt.show()
```
